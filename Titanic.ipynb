```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.feature_selection import RFE
```

### Import raw data

```python
titanic_train = pd.read_csv('data/train.csv')
titanic_test = pd.read_csv('data/test.csv')
titanic_gender= pd.read_csv('data/gender_submission.csv')
```

```python
titanic_train.head()
```

```python
print(titanic_train.shape)
print(titanic_test.shape)
```

```python
ttrain= titanic_train.drop(['Cabin', 'Name', 'Ticket', 'Age'], axis = 1)
ttrain1= ttrain.dropna()
print(ttrain1.shape)
```

```python
ttrain1.reset_index(drop=True)
```

```python
ttest= titanic_test.drop(['Cabin', 'Name', 'Ticket', 'Age'], axis = 1)
#ttest1= ttest.dropna()
ttest.reset_index(drop=True)
```

```python
Sex = {'male': 1,'female': 2}
ttrain1.Sex = [Sex[item] for item in ttrain1.Sex]
```

```python
ttrain1.head()
```

```python
ttest.Sex = [Sex[item] for item in ttest.Sex]
ttest.head()
```

```python
Embarked = {'S':1, 'Q':2, 'C': 3}
ttrain1.Embarked = [Embarked[item] for item in ttrain1.Embarked]
```

```python
ttrain1.head()
```

```python
ttest.Embarked = [Embarked[item] for item in ttest.Embarked]
```

```python
ttest.head()
```

```python
ttrain1.describe()
```

```python
ttrain1.shape, ttest.shape
```

```python
ttest.fillna(ttest['Fare'].mean(), inplace=True)
ttest.isna().sum()
```

```python
# Build your Logistic Regression model
X_train= ttrain1.drop(['Survived', 'PassengerId'], axis=1)
y_train=ttrain1['Survived']
X_test= ttest.drop(['PassengerId'], axis=1)

lr= LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_hat_train= lr.predict(X_train)
print(accuracy_score(y_train, y_hat_train))
print(confusion_matrix(y_train, y_hat_train))
```

```python
# Random Forest Classifier model
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth=2, random_state=0)
rfc.fit(X_train, y_train)
```

```python
y_pred=rfc.predict(X_train)
print(accuracy_score(y_train, y_pred))
```

```python
# Naive Bayes model
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_train)
print(accuracy_score(y_train, y_pred))
```

```python
#Choose rfc model with highest accuracy score
y_hat_test = rfc.predict(X_test)
```
